### Q3. Which player has highest strike-rate ?

### Q4.What is the favourite decision of teams who won the toss?

```python
field = bat =0
for i in range(len(mat)) :
    if mat.iloc[i].toss_decision == 'field' :
        field+=1
    elif mat.iloc[i].toss_decision == 'bat' : 
        bat+=1
matplotlib.rcParams['font.size'] = 10        
plt.bar(['FIELD','BAT'],[field,bat],color=['b','g'])
plt.xlabel(" Choice ")
plt.ylabel(" Count ")
plt.title('Favourite choice of toss winners');
```

### Explanation :
From the above graph it is clear that the Toss winning teams mostly choose the field than bat.

### Q5. Number of times the match was played between 'Chennai Super Kings' and 'Mumbai Indians' and who has the majority of wins? 

```python
matches1 = 0
chennai_wins = 0
mumbai_wins = 0
c = "Chennai Super Kings"
n = "Mumbai Indians"
x = ["Mumbai Indians","Chennai Super Kings"]
for i in range(len(mat)) :
    if mat.iloc[i].team1 in x and mat.iloc[i].team2 in x :
        matches1+=1
        if mat.iloc[i].winner == c :
            chennai_wins+=1
        if mat.iloc[i].winner == n :
            mumbai_wins+=1
print(mumbai_wins,chennai_wins,matches1)            
            
```

### Explanation :
From the above code snippet it is clear that 

28 Matches held between Chennai Super Kings and Mumbai Indians

17 Matches was won by Mumbai Indians

11 Matches was won by Chennai Super Kings

Hence Mumbai Indians has highest number of wins than Chennai Super Kings
